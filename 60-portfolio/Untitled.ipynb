```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_excel("agg_data.xlsx")
df.head()
```

## Convert to DateTime object

```python
df["date"] = pd.to_datetime(df[['year', 'month', 'day']])
first_col = df.pop("date")
df.insert(1,'date',first_col)
df = df.drop(['year', 'month', 'day'],axis=1)
df.head()
```

## Label and Tag day of the week

```python
df["day_of_week"] = df.date.dt.weekday+1
df.head()
```

```python
def fix_wknds_holidays(df, return_dates):    
    ticker_nan = df[df['ticker'].isnull()]
    #return_dates = return_df['date'].drop_duplicates().values.tolist()
    while len(ticker_nan) != 0:
        ticker_nan.date = pd.DatetimeIndex(ticker_nan['date'])-datetime.timedelta(1)

        for i in range(len(ticker_nan)):
            if ticker_nan.date.iloc[i].value in return_dates: # if new dates are in return dates then add ticker
                ticker_nan.ticker.iloc[i] = ticker_nan.tags.iloc[i]

        fixed_data = ticker_nan[ticker_nan.ticker.notnull()]

        df.loc[fixed_data.index, :] = fixed_data[:]
        ticker_nan = df[df['ticker'].isnull()]
```
