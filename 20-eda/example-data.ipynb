# drop useless columns

```python
cols = [0,2,15,16,17,18,19,20,21,22,24,25,26]
df = df.drop(df.columns[cols], axis=1)
df.head()
```

```python
df.tail()
```

# convert date stamp 

```python
date_time = df.iloc[:,0]
```

```python
date = [str(n)[0:8] for n in date_time]
date[0:5]
```

```python
import datetime
from datetime import time
from datetime import datetime as dt
date = [dt(year=int(d[0:4]), month=int(d[4:6]), day=int(d[6:8])) for d in date]
date[0:5]
```

```python
time = [str(n)[8:] for n in date_time]
time[0:5]
```

```python
time = [datetime.time(hour=int(t[0:2]), minute=int(t[2:4]), second=int(t[4:6])) for t in time]
time[0:5]
```
