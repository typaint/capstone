```python
import pandas as pd
import time
from os.path import exists
# There are 2 tables on the Wikipedia page
# we want the first table

tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
sp500_df = tables[0]
sp500_df
```

```python
tickers = sp500_df.loc[:,'Symbol']
tickers = tickers.to_list()
tickers
```

```python
from bs4 import BeautifulSoup
import requests
import re
URL = "https://www.marketcaphistory.com/?symbol=aapl"
page = requests.get(URL,headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'})
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
tables = soup.find_all('table')
table = soup.find('td', class_='infotablebox')
subtable = table.find('table')
subtable
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['date', 'mkt_cap_B'])

# Collecting Ddata
for row in subtable.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if(columns != []):
        dt = columns[0].text.strip()
        mc = columns[1].text.strip()

        df = df.append({'date': dt, 'mkt_cap_B': mc}, ignore_index=True)
df
```

```python
df.mkt_cap_B = df.mkt_cap_B.str.replace('$', '')
for i in range(1,len(df.mkt_cap_B)):
    cap = df.mkt_cap_B[i]
    if 'T' in cap:
        cap = cap.replace('T', '')
        cap = float(cap)
        cap = cap * 1000
        df.mkt_cap_B[i] = cap
    elif 'B' in cap:
        cap = cap.replace('B', '')
        cap = float(cap)
        df.mkt_cap_B[i] = cap
df.mkt_cap_B
```

```python
# convert dates to 3 columns (year, month, day)
df.year = pd.to_numeric(df.year)
df = df[df.year>=2015]
df = df.drop(columns={'change'})
# add compnay ticker
df['ticker'] = [ticker] * len(df)
df
```
