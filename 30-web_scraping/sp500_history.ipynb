```python
import pandas as pd
import time
from os.path import exists
# There are 2 tables on the Wikipedia page
# we want the first table

tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
sp500_df = tables[1]
sp500_df
```

```python
sp500_df.columns
```

```python
sp500_df.shape
```

```python
sp500_df.columns = sp500_df.columns.droplevel()
sp500_df
```

```python
sp500_df.columns.values[1:3] = ['add_ticker', 'add_sec']
sp500_df.columns.values[3:5] = ['rm_ticker', 'rm_sec']
sp500_df
```

```python
add_df = sp500_df[['Date','add_ticker', 'add_sec', 'Reason']]
add_df = add_df.drop(add_df[add_df['add_ticker'].isnull()].index,axis=0)
add_df = add_df.rename(columns={'add_ticker': 'ticker', 'add_sec': 'security'})
add_df['action'] = 'add'
add_df
```

```python
rm_df = sp500_df[['Date','rm_ticker', 'rm_sec', 'Reason']]
rm_df = rm_df.drop(rm_df[rm_df['rm_ticker'].isnull()].index,axis=0)
rm_df = rm_df.rename(columns={'rm_ticker': 'ticker', 'rm_sec': 'security'})
rm_df['action'] = 'remove'
rm_df
```

```python
frames = [rm_df, add_df]
df = pd.concat(frames).sort_index()
df
```
